# Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Util functions

```python
def drop_values_randomly(df, fraction=0.1):
    df_missing = df.copy()
    mask = np.random.rand(*df.shape) < fraction
    df_missing = df_missing.mask(mask)
    return df_missing
```

# Read data

```python

df = pd.read_csv("./data.csv")

df.head()
```

# Create copy of dataframe by randomly dropping values

```python
df_with_missing = drop_values_randomly(df, 0.1)
df_with_missing.to_csv("data_with_missing_values.csv", index=False)
```

```python
df_with_missing = pd.read_csv("data_with_missing_values.csv")
df_with_missing.head()
```

# Initial EDA

## EDA of original dataframe

### Descriptive statistics

```python
df.describe().T
```

### Histogram to understand the distribution

```python
for i in df.select_dtypes(include=["float"]).columns:
    sns.histplot(data=df, x=i)
    plt.show()
```

### Box plot to identify outliers

```python
# Exclude columns that contain only NaN values
float_columns = df.select_dtypes(include=["float"]).columns
columns_to_plot = float_columns[df[float_columns].notna().any()]

for i in columns_to_plot:
    sns.boxplot(data=df, x=i)
    plt.show()
```

### Scatter plot to identify relationship

```python
# Exclude columns that contain only NaN values
float_columns = df.select_dtypes(include=["float"]).columns
columns_to_plot = float_columns[df[float_columns].notna().any()]

for i in columns_to_plot:
    sns.scatterplot(data=df, x=i, y='diagnosis')
    plt.show()
```
